Assignment no_2. To perform data pre-processing and partitioning for a Federated Learning system using a real
dataset such as a student performance dataset (e.g., study hours, attendance, internal marks, final
score). The dataset is cleaned by handling missing values and normalization, then split into
multiple local datasets and distributed among participating devices or nodes so that each client
trains on its own private data without sharing raw information.

**IMPORT LIBRARIES**

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
```

**LOAD THE DATASET**

```python
import pandas as pd

df = pd.read_excel("student_performance.xlsx")
df.to_csv("student_performance.csv", index=False)

```

**DATA INSPECTION**

```python
print(df.head())
print(df.info())
print(df.isnull().sum())
df.fillna(df.mean(), inplace=True)
```

**SEPARATE FEATURES AND TARGET**

```python
X = df.drop(columns=["final_score"])
y = df["final_score"]
```

**NORMALIZE FEATURES**

```python
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)
```

**CONVERT BACK TO DATAFRAME**

```python
X_scaled = pd.DataFrame(X_scaled, columns=X.columns)
num_clients = 5
```

**SIMULATE FEDERATED CLIENTS**

```python
data = pd.concat([X_scaled, y.reset_index(drop=True)], axis=1)
data = data.sample(frac=1, random_state=42).reset_index(drop=True)
```

**PARTITION DATA INTO CLIENT DATASETS**

```python
client_data = np.array_split(data, num_clients)
```

**VERIFY CLIENT ISOLATION**

```python
for i, client in enumerate(client_data):
    print(f"Client {i+1} data shape: {client.shape}")
```

**STRUCTURE CLIENT DATA PROPERLY**

```python
clients = []

for client_df in client_data:
    X_client = client_df.drop(columns=["final_score"]).values
    y_client = client_df["final_score"].values
    clients.append((X_client, y_client))
```
